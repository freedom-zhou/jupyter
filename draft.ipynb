````python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# sigmoid figure
X = np.linspace(-7,7)
Y = 1 / (1 + np.e**(-X))
plt.subplot(121)
plt.grid(True)
plt.title('sigmoid 曲线')
plt.plot(X,Y)

# fibonacchi sequence
N = 10
# recursive generate fibonacchi
def fib_recur(n) -> 0:
    assert n >= 0, 'n不能小于0'
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fib_recur(n-2) + fib_recur(n-1)
    
X = np.arange(N)
Y = [fib_recur(i) for i in range(N)]
Y2 = X**2
Y3 = X * np.log2(X)
plt.subplot(122)
plt.plot(X,Y,'bo',label='斐波那契数列')
plt.plot(X,Y2,'m',label='平方阶')
plt.plot(X,Y3,'c',label='线性对数阶')

plt.title('斐波那契数列 & 平方阶 & 线性对数阶')
plt.legend()
plt.grid(True)

for ind,val in enumerate(Y):
    plt.text(ind,val+0.5,val,fontsize=12,ha='left')

plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# 准备数据
X = np.arange(1,20)
# 常数阶O(1)
Y1 = X**0 
# 对数阶O(logN)
YlogN = np.log2(X)
# 线性阶O(n)
Yn = X
# 线性对数阶O(nlogN)
YnlogN = X*np.log2(X)
# 平方阶O(n²)
Yn2 = X**2
# 立方阶O(n³)
# Yn3 = X**3
# K次方阶O(n^k)
# Yn5 = X**5
# 指数阶(2^n)
Y2n = 2**X

# 开始画图
plt.figure(figsize=(18,12))
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.plot(X, Y1, 'ro-', label="常数阶O(1)")
plt.plot(X, YlogN, 'yo-', label="对数阶O(logN)")
plt.plot(X, Yn, 'bo-', label="线性阶O(n)")
plt.plot(X, YnlogN, 'go-', label="线性对数阶O(nlogN)")
plt.plot(X, Yn2, 'co-', label="平方阶O(n²)")
# plt.plot(X, Yn3, 'mo-')
# plt.plot(X, Yn5, 'wo-')
plt.plot(X, Y2n, 'ko-', label="指数阶(2^n)")

#设置图例
plt.legend(loc='upper right')

plt.ylim(0,100)
plt.title('不同时间复杂度的对比')

plt.show()
````

为什么`all([])`的结果是True？

网上查询得知，all的原理大致如下：
```
def all(iterable):
    for x in iterable:
        if bool(x) is False:  # 只要有一个不满足,就返回False
            return False
    return True
```
① 对于all([])， 列表没有元素，也就没有不满足的，于是就返回True了  
② 对于all([[]])， 列表有1个元素，这个元素是[]，bool([])是False，所以这个元素不满足，等价于 all([False])，所以整个断言就失败了，返回False  
③ 对于all([[[]]])，列表有1个元素，这个元素是[[]]， 即有一个空列表的列表，bool([[]])是True，所以这个元素满足，等价于all([True])，所以整个断言成功，返回True  

对于②和③来说，它们都是有1个元素，区别在于元素[]和[[]]，一个是没有东西， 一个是有一个东西，虽然这东西是空列表。bool(没有东西)是False， bool(有东西)是True。
