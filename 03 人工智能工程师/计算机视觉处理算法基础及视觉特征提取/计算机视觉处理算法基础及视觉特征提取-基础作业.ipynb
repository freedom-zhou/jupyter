#### 1. 画图解释图像卷积滤波的基本原理，并进一步简述常见的图像平滑滤波算法。 
* 卷积滤波的基本原理：滤波即卷积，由逐点乘积后累加得到
![image.png](./滤波原理.png "卷积即逐点乘积后累加")

* 常见的图像平滑滤波算法有：
    - 平均滤波
    - 加权平均滤波(以高斯滤波为代表)
    - 中值滤波(对椒盐噪声很有效)
    
#### 2. 简述边缘检测的基本原理，以及Sobel、LoG和Canny算子的原理差异。 
* 边缘检测的基本原理：使用微分或差分找到图像的变化剧烈的点（一重微分的峰值或者二重微分的过零点）
* Sobel算子的特点是简单，常用，原理是对图像的一阶差分
* Log算子是为了解决Laplase算子噪声敏感的问题而在其基础上加了高斯平滑，缺点是存在边缘断裂，原理是对图像的二阶差分+高斯平滑
* Canny算子最复杂，但同时解决了其它算子的三大缺点，即噪声敏感、边缘断裂、边缘虚检，原理是对图像的一阶差分+高斯平滑+非极大值抑制+双阈值（边缘连接）

#### 3. 简述图像灰度直方图的基本概念，及使用大津算法进行图像分割的基本原理。 
* 灰度直方图：以图像灰度值为横坐标，灰度值出现的次数或频率为纵坐标的统计直方图
* 大津算法基本原理：遍历灰度所有可能的取值，找到令类间方差最大的值作为阈值来分割
