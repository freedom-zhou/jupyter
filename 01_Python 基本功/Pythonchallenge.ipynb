# 第一关（关键词：字符串翻译str.maketrans,str.translate）

```python
import re
import urllib
import string

url = 'http://www.pythonchallenge.com/pc/def/map.html'
regex = re.compile(r'<font color="#f000f0">(.*?)</tr>', re.S)

# 爬取指定网页源码，并提取关键文本
response = urllib.request.urlopen(url)
html = response.read().decode('utf-8')
text = regex.search(html).group(1)
print(text)

# 变换字符串
chars = string.ascii_lowercase
tran_map = str.maketrans(chars, chars[2:]+chars[:2])
text=text.translate(tran_map)
print(text)
```

# 第六关（关键词: 操作压缩文件 zipfile）

```python
import re
import zipfile

path = './data/channel.zip'
f_name = 'readme.txt'
            
comments = []
with zipfile.ZipFile(path, 'r') as z:
    for i in range(920):
        # 获取文件comment
        comments.append(z.getinfo(f_name).comment.decode('utf-8'))
        
        # 读取文件内容，获取下一个文件名
        text = z.read(f_name)
        num_str = ''.join(re.findall(r' (\d+)',text.decode('utf-8')))
        if num_str.isdigit():
            f_name = num_str + '.txt'
        else:
            break
            
print(''.join(comments))
```

# 第十关（找规律）

已知 a = [1, 11, 21, 1211, 111221, ...]  
问：len(a[30]) = ?

```python
# 创建生成器函数
def geta(n: int):
    a = '1'
    count = 0
    while True:
        if count >= n:
            break
        yield a
        
        count += 1
        a += '?'
        nexta = []
        counti, i = 0, a[0]
        for x in a:
            if x == i:
                counti += 1
            else:
                nexta.append(str(counti))
                nexta.append(i)
                counti = 1
                i = x
        a = ''.join(nexta)
        
a30 = [i for i in geta(31)][30]
print(len(a30))
        
```

# 第十一关（关键词：图像拆分 PIL）

```python
import requests
import PIL.Image

# img_url = "http://www.pythonchallenge.com/pc/return/cave.jpg"

# # 爬取图片 (需要登录，爬取失败)
# response = requests.get(img_url)
# byte_img = response.content
# # 保存图片
# with open("data/cave.jpg", 'wb') as f:
#     f.write(byte_img)

# 打开图片
img0 = Image.open("data/cave.jpg")
# 获取图片信息
mode = img0.mode
w, h = img0.size

# 创建空白图片
img1 = Image.new(mode, (w, h), '#FFF')
img2 = Image.new(mode, (w, h), '#FFF')
img3 = Image.new(mode, (w, h), '#FFF')
img4 = Image.new(mode, (w, h), '#FFF')

# 加载图片矩阵
img_array1 = img1.load()
img_array2 = img2.load()
img_array3 = img3.load()
img_array4 = img4.load()

for x in range(w):
    for y in range(h):
        if x % 2 == 0 and y % 2 == 0:
            img_array1[x/2, y/2] = img0.getpixel((x, y))
        elif x % 2 != 0 and y % 2 != 0:
            img_array2[x/2, y/2] = img0.getpixel((x, y))
        elif x % 2 != 0 and y % 2 == 0:
            img_array3[x/2, y/2] = img0.getpixel((x, y))
        elif x % 2 == 0 and y % 2 != 0:
            img_array4[x/2, y/2] = img0.getpixel((x, y))
            
img1.save('data/cave1.jpeg', 'jpeg')
img2.save('data/cave2.jpeg', 'jpeg')
img3.save('data/cave3.jpeg', 'jpeg')
img4.save('data/cave4.jpeg', 'jpeg')
```
