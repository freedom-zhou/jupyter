# urllib

## get请求

```python
from urllib import request

baseUrl = 'http://www.baidu.com'

# 发起请求并保存获取到的响应
response = request.urlopen(baseUrl)
# 读取数据
bytes_html = response.read()
# 解码二进制数据
html = bytes_html.decode('utf-8')

# 写入文件
with open(f'./data/baidu_urllib.html', 'w+', encoding='utf-8') as file_w:
    file_w.write(html)
```

```python
print(type(bytes_html))
print(type(response))
print( dir(response))
```

## post请求

```python
from urllib import request, parse, error

url1 = 'http://httpbin.org/post' # 专门用来测get、post、put、delete等请求的网站
url2 = 'http://www.baidu.com' 
url3 = 'https://www.douban.com'  # 豆瓣做了反爬

data = bytes(parse.urlencode({'hello':'world'}), encoding='utf-8')
err = None

# post请求需要传data
try:
    response = request.urlopen(url1, data=data, timeout=0.1) # 设置超时时间
    print('服务器响应数据：')
    print(response.read().decode('utf-8'))
    print('状态码：')
    print(response.status)
    print('请求头：')
    print(response.getheaders())
except error.URLError as e:
    err = e
    print(e.reason)
except Exception as e:
    err = e
    print('reason:')
    print(e.reason)
except:
    print('Unknow error')
```

```python
print(type(err))
print( dir(err))
print(     err.reason )
# print(     err.code )
```

## 避免反爬

### 加请求头，伪装成浏览器

```python
from urllib import request, parse, error

url1 = 'http://httpbin.org/post' # 专门用来测get、post、put、delete等请求的网站
url2 = 'http://www.baidu.com' 
url3 = 'https://www.douban.com'  # 豆瓣做了反爬

# data = bytes(parse.urlencode({'name':'zrd'}), encoding='utf-8')
# 加入浏览器的头部信息，这里伪装成谷歌浏览器
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36'
}
# req = request.Request(url=url1, data=data, headers=headers, method='POST') 
req = request.Request(url=url3, headers=headers) 

try: 
    response = request.urlopen(req, timeout=3) # 设置超时时间
    print('服务器响应数据：')
    print(response.read().decode('utf-8'))
    print('状态码：')
    print(response.status)
    print('请求头：')
    print(response.getheaders())
except error.URLError as e:
    print(e.reason)
```

# requests

```python
import requests

baseUrl = 'http://www.baidu.com'

# response = requests.request('get', baseUrl)
response = requests.get(baseUrl) # 和上条代码等价

# 直接获取text会乱码
html = response.text

# 解决乱码 方法1
# response.encoding = 'utf-8' # 先设置编码格式为utf-8
# html = response.text # 再获取text

# 解决乱码 方法2
bytes_html = response.content # 获取bytes型
html = bytes_html.decode('utf-8') # 再转为str型

# 写入文件
with open(f'./data/baidu_requests.html', 'w+', encoding='utf-8') as file_w:
    file_w.write(html)
```

## 避免反扒

### 加请求头

```python
import requests

url = 'https://www.douban.com'  # 豆瓣做了反爬

# 加入浏览器的头部信息，这里伪装成谷歌浏览器
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36'
}

response = requests.get(url,headers=headers)

print(response.url)
print(response.status_code)
print(response.request.headers)
print("="*50)
print(response.content.decode("utf-8"))
```
