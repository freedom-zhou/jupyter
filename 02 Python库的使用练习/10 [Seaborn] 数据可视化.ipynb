# barplot 柱状图，heatmap 热力图

**flight_delays.csv 数据集**每个cell代表每家航空公司在2015年每个月的平均延迟时间（分钟），负数代表航班提前到达，例如美国航空(AA)在一月平均延误7分钟

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

flight_data = pd.read_csv('data/flight_delays.csv', index_col = 'Month')
flight_data
```

## 柱状图 (seaborn.barplot)

```python
# 设置图像大小
plt.figure(figsize=(10,6))

# 设置标题
plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')

# 设置y轴名称
plt.ylabel('Arrival delay (in minutes)')

# 创建柱状图
sns.barplot(x = flight_data.index, y = flight_data.NK)
```

## 热力图 (seaborn.heatmap)

```python
# 设置图片大小
plt.figure(figsize=(16,8))

# 设置标题
plt.title('Average Arrival Delay for Each Airline, by Month')

# 设置x轴名称
plt.xlabel('Airline')
# 创建热力图
sns.heatmap(data = flight_data, annot = True)
```

### heatmap 参数

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
%matplotlib inline

#  mask : 为True的元素对应位置不会画出来（mask面具的意义）
#  cmap: 颜色设置
#  square: （True）代表行列长度一致，且绘制的每个也是方格
#  annot ： 在格内显示数据
#  fmt ：数据格式

# 设置常量
PARAM_N = 4 # 要对比的参数个数

# 创建数据
# temp = list(range(10))*2
# 这里必须设置为float型，否则设置mask时会出错
temp = list(np.arange(10.0))*2
data = [temp[i:i+10] for i in range(9,-1,-1)]
data = pd.DataFrame(data)

# 创建图片
figure = plt.figure(figsize=(15,5*PARAM_N))
# figure = plt.figure()

# Parameter annot=True
ax = figure.add_subplot(PARAM_N,2,1)
ax.set_title('annot=True')
sb.heatmap(data, annot=True)

# Parameter annot=False
ax = figure.add_subplot(PARAM_N,2,2)
ax.set_title('annot=False')
sb.heatmap(data, annot=False)
#---------------------------------------------

# Parameter square=True
ax = figure.add_subplot(PARAM_N,2,3)
ax.set_title('square=True')
sb.heatmap(data, square=True, ax=ax)

# Parameter square=False
ax = figure.add_subplot(PARAM_N,2,4)
ax.set_title('square=False')
sb.heatmap(data, square=False, ax=ax)
#---------------------------------------------

# Parameter cbar=True
ax = figure.add_subplot(PARAM_N,2,5)
ax.set_title('cbar=True')
sb.heatmap(data, cbar=True, ax=ax)

# Parameter cbar=False
ax = figure.add_subplot(PARAM_N,2,6)
ax.set_title('cbar=False')
sb.heatmap(data, cbar=False, ax=ax)
#---------------------------------------------

# Parameter mask=None
ax = figure.add_subplot(PARAM_N,2,7)
ax.set_title('mask=None')
sb.heatmap(data, mask=None, ax=ax)

# Parameter mask=mask
# mask初始化为和data一样大的 全0(False)矩阵
mask = np.zeros_like(data, dtype=np.bool)
# 将mask右上三角(列号》=行号)设置为True
# mask[np.triu_indices_from(mask)] = True
mask[::2,::2] = True
ax = figure.add_subplot(PARAM_N,2,8)
ax.set_title('mask=mask')
sb.heatmap(data, mask=mask, ax=ax)
#---------------------------------------------
```

# 直方图、密度图、二维密度图

这次我们会使用Iris数据集，Iris是常用的分类实验数据集，Iris也称鸢尾花卉数据集，是一类多重变量分析的数据集。数据集包含150个数据样本，分为3类，每类50个数据，每个数据包含4个属性。可通过花萼长度，花萼宽度，花瓣长度，花瓣宽度4个属性预测鸢尾花卉属于（Setosa，Versicolour，Virginica）三个种类中的哪一类。


iris以鸢尾花的特征作为数据来源，常用在分类操作中。该数据集由3种不同类型的鸢尾花的各50个样本数据构成。其中的一个种类与另外两个种类是线性可分离的，后两个种类是非线性可分离的。
该数据集包含了4个属性：



Sepal.Length（花萼长度），单位是cm;

Sepal.Width（花萼宽度），单位是cm;

Petal.Length（花瓣长度），单位是cm;

Petal.Width（花瓣宽度），单位是cm;


种类：Iris Setosa（山鸢尾）、Iris Versicolour（杂色鸢尾），以及Iris Virginica（维吉尼亚鸢尾）。

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
iris_data = pd.read_csv('data/iris.csv')
iris_data.head()
```

```python
help(sb.distplot)
```

## 直方图 （seaborn.distplot）

```python
sb.distplot(a = iris_data['Petal Length (cm)'], kde = False) # kde: kernel density estimate
```

## 箱线图 （seaborn.boxplot）

```python
sb.boxplot(x='Species', y='Petal Length (cm)', data=iris_data)
```

## 提琴图 (seaborn.violinplot)

```python
sb.violinplot(x='Species', y='Petal Length (cm)', data=iris_data)
```

## 密度图 (seaborn.kdeplot)

```python
# kde: kernel density estimate
sb.kdeplot(data = iris_data['Petal Length (cm)'], shade=True)
```

## 二维密度图 (seaborn.jointplot)

```python
sb.jointplot(x = iris_data['Petal Length (cm)'], y = iris_data['Sepal Width (cm)'], kind = 'kde')
```
