## matplotlib.pyplot.figure()函数 创建图片

matplotlib.pyplot.figure(num=None, figsize=None, dpi=None, facecolor=None, edgecolor=None, frameon=True, FigureClass=<class 'matplotlib.figure.Figure'>, clear=False, **kwargs)  
returns：返回一个图片 

参数详解：

* num : integer or string, optional, default: None
    * 默认None则创建一个图形，图形标号自动递增；
    * 如果提供了一个数，并且与已有id重合，激活它并返回索引；
    * 如果提供的数不存在，创建并返回。
    * 如果提供的是str，就返回到窗口标题上。
    * 提供的num参数存放在figure对象的number属性里

* figsize : (float, float), optional, default: None
英寸单位的宽和高，默认为 rcParams["figure.figsize"] = [6.4, 4.8].

* dpi : integer, optional, default: None
图像的分辨率，默认 rcParams["figure.dpi"] = 100.

* facecolor :
背景颜色，默认 rcParams["figure.facecolor"] = 'w'.

* edgecolor :
边的颜色，默认 rcParams["figure.edgecolor"] = 'w'.

* frameon : bool, optional, default: True
如果设为False, 禁止绘制图形边框.

* FigureClass : subclass of Figure
Optionally use a custom Figure instance.

* clear : bool, optional, default: False
如果是True，并且图形已经存在，则清楚该图形

**如果创建了很多张图片，一定要采用pyplot.close()关闭不用的图片，避免内存占用过大。**

## matplotlib.pyplot.plot()函数 绘制线形图

```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib
%matplotlib inline

# 最简单的例子
x = np.arange(1,11)
y = x * 2 + 6
plt.plot(x,y,'b') # blue
plt.show()
```

```python
# 参数  [color][marker][line]
# 限制x轴的范围 plt.xlim
# 限制y轴的范围 plt.ylim

plt.rcParams['figure.figsize'] = (18,6)
plt.rcParams['font.sans-serif'] = ['SimHei']
x1 = np.arange(0,20,2)
x2 = x1+20
x3 = x2+20
y1 = x1**2
y2 = np.array([0,100,0,100,0,100,0,100,0,100])
y3 = 3*x3 + 10

plt.plot(x1, y1, 'r') # red
plt.plot(x1, y1+30, 'r+') # red
plt.plot(x1, y1+60, 'r--') # red [color][marker][line]

plt.plot(x2, y2, 'b') # blue
plt.plot(x2, y2+30, 'b*') # blue
plt.plot(x2, y2+60, 'b-') # blue
plt.plot(x2, y2+90, 'b--') # blue
plt.plot(x2, y2+120, 'b|') # blue
plt.plot(x2, y2+150, 'b|-') # blue

plt.plot(x3, y3, 'g') # green
plt.plot(x3, y3+30, 'go') # green
plt.plot(x3, y3+60, 'g|-') # green

# 显示图例
plt.legend('r r+ r-- b b* b- b-- b| b|- g go g|-'.split(),loc='upper right')

plt.xlim([0,65]) # 限制x轴的范围
plt.ylim([0,400])
plt.title('测试不同的marker和line style')
plt.grid(ls="--") # 显示网格
plt.show()
```

## 绘制子图

### 方式1 （matplotlib.pyplot.subplot()函数）

```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib

# 选择绘图模式
%matplotlib inline
# %matplotlib notebook
# %matplotlib auto

# 计算正弦和余弦曲线上的点的 x 和 y 坐标 
x = np.arange(0,  4  * np.pi,  0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  

# 设置图片大小
plt.rcParams['figure.figsize'] = (10,8)
# 使负号正常显示
plt.rcParams['axes.unicode_minus'] = False

# 建立第一个subplot 网格，高为 2，宽为 1 ,第1个图像
plt.subplot(2,  1,  1)  
# 绘制第一个图像 
plt.plot(x, y_sin) 
plt.title('Sine')  

# 建立第二个subplot 网格，并绘制第二个图像
plt.subplot(212)
plt.plot(x, y_cos) 
plt.title('Cosine')
# 调整图像间隔
plt.tight_layout()
# 展示图像
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib

# 选择绘图模式
%matplotlib inline
# %matplotlib notebook
# %matplotlib auto

fig = plt.figure(figsize=(12,6))

shape=['.','o','v','>','<','8','s','*']
for j in range(8):
    x=[i for i in range(6)]
    y=[i**2 for i in range(6)]
    ax = fig.add_subplot(241+j)
    ax.scatter(x,y,c='r',marker=shape[j])
#     ax.bar(x,y)
    ax.set_title('No.'+str(j))
plt.tight_layout()
plt.show()
```

### 调整子图间隔 (matplotlib.pyplot.subplots_adjust())

```python
#subplots_adjust(left=None, bottom=None, right=None, top=None,wspace=None, hspace=None)
# 参数说明：
# top、bottom、left、right：整个图距离上下左右边框的距离
# wspace、hspace：这个才是调整各个子图之间的间距
# wspace：调整子图之间的横向间距
# hspace：调整子图之间纵向间距

import numpy as np
import matplotlib.pyplot as plt
%matplotlib auto

fig, axes = plt.subplots(2, 2, figsize=(3,3))

for i in np.arange(0.1,0.8,0.1):
    plt.subplots_adjust(wspace=i*3, hspace=i*3)
    plt.pause(1)

# plt.cla()

for i in np.arange(0.1,0.8,0.1):
    plt.subplots_adjust(left=i, bottom=i)
    plt.pause(1)
    
# plt.cla()
# plt.clf()
# plt.close()
plt.show()
```

### 画圆 (三种方法：标准方程，参数方程，matplotlib自带函数库)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# 设置常量
N = 3 # 子图数
WIDTH = 3 # 单个子图的宽度
COLOR = 'r' # 圆的颜色

# 设置plt参数
plt.rcParams['font.sans-serif'] = ['SimHei'] # 中文正常显示
plt.rcParams['axes.unicode_minus'] = False # 负数正常显示

# 创建画板和子图
fig,axes = plt.subplots(1, N, figsize=((WIDTH+2)*N, WIDTH+1))

# 设置圆形的参数
a = 0.0
b = 0.0
r = 10.0


# 1.标准方程 画圆
# 设置x，y
x = np.arange(a-r, a+r+0.1, 0.1)
y = b + np.sqrt(r**2 - (x-a)**2)
# 开始画图
axes[0].plot(x, y, )
axes[0].plot(x, -y, )
# 设置标题
axes[0].set_title('标准方程')
# 其它设置
axes[0].grid(ls="--") # 显示网格

# 2.参数方程 画圆
# 设置x，y
theta = np.arange(0, np.pi*2, 0.01)
x = a + r * np.cos(theta)
y = b + r * np.sin(theta)
# 开始画图
axes[1].plot(x, y, COLOR)
# 设置标题
axes[1].set_title('参数方程')
# 其它设置
axes[1].grid(ls="--") # 显示网格

# 3.调用matplotlib.patches包函数绘制圆形Circle和椭圆Ellipse
from matplotlib.patches import Ellipse, Circle
cir1 = Circle(xy = (0.0, 0.0), radius=r, alpha=0.5, facecolor='green')
ell1 = Ellipse(xy = (0.0, 0.0), height = 2*r, width = r, angle = 45.0, facecolor= 'green', alpha=0.3)
axes[2].add_patch(ell1)
axes[2].add_patch(cir1)
# x, y = 0, 0
# axes[2].plot(x, y, 'ro')
# 设置标题
axes[2].set_title('matplotlib.patches.Circle')
# 其它设置
axes[2].grid(ls="--") # 显示网格

# 调整图像
plt.xlim(a-r,a+r)
plt.ylim(a-r,a+r)
plt.subplots_adjust(wspace=0.3,hspace=0.3) # 调整子图横向纵向间隔
# plt.tight_layout()
# plt.axis('square') # 强制使x，y轴长度相等
# plt.axis('equal') # 使x，y轴尺度相同
# plt.axis('scaled')
# plt.axis('off')   #turns off the axis lines and labels
plt.show()
```
