#### 3. 什么是光通量和辐照度？说明几个常见光源的光通量，以及几个常见照明环境的辐照度。 
* 光通量：指人眼所能感受到的辐射功率，它等于单位时间内某一波段的辐射能量和该波段的相对视见率的乘积，以符号Φ表示，单位是lm(流明)
* 辐照度：指投射到一平表面上的辐射通量密度，指到达一表平面上，单位时间，单位面积上的辐射能，以符号E表示，常用单位是lux(勒克斯)
* 常见光源的光通量
    - 太阳：3.566×1028 lm
    - 烛光：12.56 lm
* 常见照明环境的辐照度
    - 黑夜：0.001~0.02 lux
    - 阴天室内：5~50 lux
    - 晴天阳光直射：100000 lux

#### 5. 说明彩色图像传感器及γ校正的基本原理。 
* 彩色图像传感器的原理：  
彩色图像传感器包括上下三层结构，各有很多小格，最底层的传感器层敏感灰度信息；中间的薄膜层的小格有红绿蓝三种颜色，不同的颜色敏感格自的颜色；顶层是微透镜层，用以汇聚中间层透出的颜色光线
* γ校正的基本原理：  
    由于人眼的视觉感应和照明值并非是线性关系，因此要采用指数形式的γ校正。具体应用中，在图像传感器采集图像后需要进行γ校正后输出影像，已复核人的视觉习惯，存储时又要经过逆变换还原回原有的RGB值
